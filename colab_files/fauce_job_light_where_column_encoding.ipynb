```python

import pandas as pd

# Read the keyword.csv file into a Pandas DataFrame
df = pd.read_csv('movie_keyword.csv', on_bad_lines='skip', low_memory=False)
```

find min, max for all the columns in the table.

```python

def find_min_max(df, table_name):
  min_max = {}
  for column in df:
    min_value = df[column].min()
    max_value = df[column].max()
    min_max[table_name + '.' + column] = [min_value, max_value]
  return min_max

min_max = find_min_max(df, 'movie_keyword')

print(min_max)
```

fill NA values for columns and encode string column using ordinal encoder

```python
title_df.fillna({"production_year": 0, "episode_of_id": -1, "season_nr": -1, "episode_nr": -1, "imdb_id": -1}, inplace=True)
title_df['title'].fillna('', inplace=True)
title_df['imdb_index'].fillna('', inplace=True)
title_df['phonetic_code'].fillna('', inplace=True)
title_df['series_years'].fillna('', inplace=True)
title_cols = ['title', 'imdb_index', 'phonetic_code', 'series_years', 'md5sum']
title_ordinal_encoder = OrdinalEncoder(unknown_value=-1, handle_unknown='use_encoded_value')
title_df[title_cols] = title_ordinal_encoder.fit_transform(title_df[title_cols])
```

```python
min_max_dict = {}

# Iterate over all the dataframes
for (_df, _name) in [(df, "movie_keyword"), (movie_companies_df, "movie_company"), (movie_info_df, "movie_info"), (movie_info_idx_df, "movie_info_idx"), (cast_info_df, "cast_info"), (title_df, "title")]:
  # Find the min and max values for each column in the dataframe
  min_max = find_min_max(_df, _name)
  # Update the dictionary with the min and max values
  min_max_dict.update(min_max)

# Print the dictionary of min and max values
print(min_max_dict)
```

encode predicates for each where clause for each query

```python
import numpy as np
encoded_column_values = []

for item in where_clauses:
  table_names = []
  encoding_dict = min_max_dict.copy()

  for i in range(0, len(item), 3):
    table_names.append(item[i].split('.')[0])
    col_value = item[i+2]

    col_name = item[i].split('.')[1]
    if col_name in title_cols:
      # select which column
      test_data = [min_reverse_encoding_dict[title] for title in title_cols]
      test_data[title_cols_dict[col_name]] = col_value
      encoding = title_ordinal_encoder.transform([test_data])
      col_value = encoding[0][title_cols_dict[col_name]]

    elif item[i] == "movie_info_idx.info":
        col_value = movie_info_idx_encoder.transform([[col_value]])[0]
    elif item[i] == "movie_info.info":
        col_value = movie_info_encoder.transform([[col_value]])[0]
    elif item[i] == "movie_company.note":
        col_value = encoder.transform([[col_value]])[0]
    else:
      print(item[i], col_value)
      col_value = float(col_value)

    if item[i+1] == "=":
      encoding_dict.update({item[i]: [col_value, col_value]})
    elif item[i+1] == ">" or item[i+1] == ">=":
      encoding_dict.update({item[i]: [col_value, encoding_dict[item[i]][1]]})
    elif item[i+1] == "<" or item[i+1] == "<=":
      encoding_dict.update({item[i]: [encoding_dict[item[i]][0], col_value]})


  for encoding_dict_item in encoding_dict:
    if encoding_dict_item.split('.')[0] in table_names:
      continue
    else:
      encoding_dict.update({encoding_dict_item: [0, 0]})
  print("*********")
  print(item)
  print(encoding_dict)
  print("*********")
  item_to_append = []
  for key in min_max_dict_keys:
    item_to_append.append(encoding_dict[key])
  encoded_column_values.append(np.array(item_to_append).flatten().tolist())
```

```python
import pandas as pd

# Create a DataFrame from the flattened encoded column values
encoded_columns_df = pd.DataFrame(encoded_column_values)

# Print the DataFrame
print(encoded_columns_df)

# Save the DataFrame to a CSV file
encoded_columns_df.to_csv('encoded_column_values.csv', index=False)
```
