```python
import pandas as pd
global_column_names = pd.read_csv("global_column_names.txt", delimiter=" ", header=None)
global_column_names
```

```python
# prompt: make a dictionary from global_column_names with values of column 0 as keys and values of column 1 as values

global_column_dict = dict(zip(global_column_names.iloc[:, 0], global_column_names.iloc[:, 1]))
global_column_dict
```

```python
# prompt: for column 2 of df, split the value by , and then get values starting from the first one with the gap of 2 for each row

def extract_values(row):
  values = row[2].split(",")
  extracted_values = [values[i] for i in range(0, len(values), 3)]
  return extracted_values

column_names_in_where = df.apply(extract_values, axis=1)
```

```python
# prompt: for each item of column_names_in_where use global_column_dict to get the values

def get_values_from_dict(column_names):
  values = []
  for name in column_names:
    if name in global_column_dict:
      values.append(global_column_dict.get(name))
    else:
      print(f"Value '{name}' not found in global_column_dict.")
      values.append(None)
  return values

where_column_names_in_numbers = column_names_in_where.apply(get_values_from_dict)
```

```python
where_column_names_in_numbers
```

```python
import pandas as pd
column_names_in_numbers = []
for row in column_in_join:
  values = []
  for name in row:
    if name in global_column_dict:
      values.append(global_column_dict.get(name))
    else:
      print(f"Value '{name}' not found in global_column_dict.")
      values.append(None)
  values = list(np.unique(values))
  column_names_in_numbers.append(values)

column_names_in_numbers

```

```python
final_col_list = []
for i in range(len(column_names_in_numbers)):
  final_col_list.append(column_names_in_numbers[i] + where_column_names_in_numbers[i])

final_col_list
```

```python
zero_encoding_mapping = {key: 0 for key in column_encoding_mapping}
```

```python
encoded_col_list = []
for col_list in final_col_list:
    zero_encoding_mapping = {key: 0 for key in column_encoding_mapping}
    for item in col_list:
        if item in column_encoding_mapping:
            zero_encoding_mapping[item] = column_encoding_mapping[item]
    encoded_col_list.append(list(zero_encoding_mapping.values()))
encoded_col_list
```

```python
def get_table_numbers_for_names(names):
  table_numbers = []
  for name in names:
    table_numbers.append(reverse_dict_from_file[name])
  return table_numbers

join_table_numbers = df[0].apply(lambda x: get_table_numbers_for_names(x.split(",")))
join_table_numbers
```

```python
def get_encoding(table_names):
  encoding = []
  table_numbers = join_table_numbers[0]
  for name in table_names:
    table_numbers.append(reverse_dict_from_file[name])
  for index, row in join_subgraphs.iterrows():
    join_subgraph = row[0]
    if set(table_numbers).issubset(set(join_subgraph)):
      encoding.append(row[1])
    else:
      encoding.append(0.0)
  return encoding

join_table_name_encoding = df[0].apply(lambda x: get_encoding(x.split(",")))
join_table_name_encoding
```
